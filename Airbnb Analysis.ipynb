```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df= pd.read_csv('AB_NYC_2019.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['name'])
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['host_name'])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['last_review']=pd.to_datetime(df['last_review'])
df['last_review_year']=df['last_review'].dt.year
```

```python
df['last_review_month']=df['last_review'].dt.month
```

```python
df.head()
```

```python
df=df.drop('last_review',axis=1)
```

```python
df['last_review_year'].fillna(df['last_review_year'].mean().round(0), inplace=True)
```

```python
df['last_review_year'].median()
```

```python
df.head()
```

```python
df['last_review_month'].fillna(df['last_review_month'].mean().round(0), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['last_review_year']=df['last_review_year'].astype(int)
```

```python
df['last_review_month']=df['last_review_month'].astype(int)
```
